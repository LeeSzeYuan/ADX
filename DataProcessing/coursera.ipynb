```python
import numpy as np
import pandas as pd
df = pd.read_csv("../input/100k-courseras-course-reviews-dataset/reviews.csv")
```

```python
df.drop(columns=['Id'],inplace=True)
df.head()
```

```python
df.isna().any()
```

```python
from textblob import TextBlob
df['sentiment'] = df.Review.apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

```python
df.head(20)
df.sentiment.describe()
```

```python
df['type'] = ['Good' if x >0.3 else 'Bad' if x < -0.3 else 'neutral' for x in df['sentiment']] 
```

```python
df.head(30)
```

```python
df.to_csv("coursera.csv")
```

```python
from gensim.summarization import keywords
```

```python
print(keywords(df.Review[900], words=1, lemmatize = True))
```

```python
df['keyword'] = [keywords(x, words=1, lemmatize = True) for x in df['Review']] 
```
